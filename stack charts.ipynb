```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv(r"C:\Users\R A H U L\Desktop\stack chart.csv")
```

```python
data
```

```python
colors = ["#011638", "#7e2987", "#ef2026"]
plt.figure(figsize=(12,6))
plt.stackplot(data["Year"], data["Gas"], data["Petrol"], data["Diesel"], colors = colors)
plt.xticks(data["Year"], rotation = "45")
plt.show()
```

```python
# In the above chart we can see white color between each stacks, which can be removed by setting edgecolor to none

colors = ["#011638", "#7e2987", "#ef2026"]
labels = ["Gas","Petrol","Diesel"]
plt.figure(figsize=(12,6))
plt.stackplot(data["Year"], data["Gas"], data["Petrol"], data["Diesel"], colors = colors, edgecolor = 'none')
plt.xticks(data["Year"], rotation = "45")
plt.legend(labels = labels)
plt.show()


#Changing the legend position
```

```python
#Changing the legend position

colors = ["#011638", "#7e2987", "#ef2026"]
labels = ["Gas","Petrol","Diesel"]
plt.figure(figsize=(12,6))
plt.stackplot(data["Year"], data["Gas"], data["Petrol"], data["Diesel"], colors = colors, edgecolor = 'none')
plt.xticks(data["Year"], rotation = "45")
plt.legend(labels = labels, loc = "upper left")
plt.title("Popularity of engine fuel types(1982 - 2016)", fontsize = 14, weight = "bold")
plt.ylabel("Number of cars", fontsize = 13)
plt.show()
```
