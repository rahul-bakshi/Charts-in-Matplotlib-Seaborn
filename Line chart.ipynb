```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv(r"C:\Users\R A H U L\Desktop\line_data.csv")
data
```

```python
data["new_date"] = pd.to_datetime(data["Date"])
```

```python
data["new_date"]
```

```python
labels = ["S&P 500", "FTSE"]
plt.figure(figsize = (20,8))
plt.plot(data["new_date"],data["GSPC500"])
plt.plot(data["new_date"],data["FTSE100"])
plt.legend(labels = labels, fontsize = "large")
plt.xlabel("Date")
plt.ylabel("Returns")
plt.title("S&P v/s FTSE Returns(2000 - 2010)", fontsize=16, fontweight="bold")
plt.show()
```

```python
# Since we have a very large dataset to get any meaningful outcome, so we plot Q3 and Q4 of 2008

data1 = data[(data.new_date >= '2008-07-01') & (data.new_date <= '2008-12-31')]
data1
```

```python
labels = ["S&P 500", "FTSE"]
plt.figure(figsize = (20,8))
plt.plot(data1["new_date"],data1["GSPC500"], color = 'midnightblue')
plt.plot(data1["new_date"],data1["FTSE100"], color = 'crimson')
plt.legend(labels = labels, fontsize = "large")
plt.xlabel("Date",fontsize = "13", fontweight = "bold")
plt.ylabel("Returns",fontsize = "13", fontweight = "bold")
plt.title("S&P v/s FTSE Returns(H2 2008)", fontsize=18, fontweight="bold")
plt.show()
```
