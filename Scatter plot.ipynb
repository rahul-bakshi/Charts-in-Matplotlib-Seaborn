```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"C:\Users\R A H U L\Desktop\scatter_data.csv")
```

```python
df
```

# With Matplotlib

```python
plt.figure(figsize = (12,8))
plt.scatter(df['Area (ft.)'],df['Price'])
plt.show()
```

```python
#To avoid overplotting we can add transparency level by setting the alpha value(0 to 1)
```

```python
plt.figure(figsize = (12,8))
plt.scatter(df['Area (ft.)'],
            df['Price'], alpha = 0.6, 
            c=df['Building Type'], #c -> color coding(output is default color)
            cmap = 'viridis') #mapping the default color to a different color(like 'mako','viridis'..etc)

plt.show()
```

```python
plt.figure(figsize = (12,8))
scatter = plt.scatter(df['Area (ft.)'],
            df['Price'], alpha = 0.6, c=df['Building Type'], cmap = 'viridis')
plt.legend(*scatter.legend_elements(), #pointer to the legend elements of the scatter
           loc = 'upper left', 
           title = "Builidng Type") 
plt.title("Relationship between Area and Price",fontsize=14,fontweight="bold")
plt.xlabel("Area in sq.ft")
plt.ylabel("Price($)")
plt.show()
```

# With Seaborn

```python
plt.figure(figsize = (12,8))

sns.scatterplot(df['Price'],
                df['Area (ft.)'],
                hue=df['Building Type'],
                palette=['black','darkblue','purple','pink','white'],
                         s=100)
plt.title("Relationship between Area and Price",fontsize=14,fontweight="bold")
plt.xlabel("Area in sq.ft")
plt.ylabel("Price($)")
plt.show()
            
```
