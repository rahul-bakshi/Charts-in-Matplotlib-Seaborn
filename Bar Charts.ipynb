# After importing seaborn

```python
plt.bar(x=data["Brand"], height=data["Cars Listings"])
plt.xticks(rotation = 45)
plt.show
```

```python
plt.figure(figsize = (9,6))
plt.bar(x=data["Brand"], 
        height=data["Cars Listings"], 
        color = "g")
plt.xticks(rotation = 45)
plt.show
```

```python
plt.figure(figsize = (9,6))
plt.bar(x=data["Brand"], 
        height=data["Cars Listings"], 
        color = "midnightblue")
plt.xticks(rotation = 45,fontsize = 13)
plt.yticks(fontsize = 13)
plt.ylabel("Number of Listings",fontsize = 13,fontweight = "bold")
plt.title("Car listings by brand",fontsize=16,fontweight="bold")
plt.show
```
