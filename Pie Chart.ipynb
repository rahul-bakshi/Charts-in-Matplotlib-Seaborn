```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv(r"C:\Users\R A H U L\Desktop\pie_chart_data.csv")
data
```

```python
plt.figure(figsize = (10,6))
plt.pie(data["Number of Cars"], 
        labels = data["Engine Fuel Type"],
        autopct = '%.2f%%',
        textprops = {'size' : 'x-large','fontweight':'bold','rotation':'35','color':'black'})
plt.title("Cars by engine fuel type",fontsize = 18,fontweight="bold")
plt.legend()
plt.show()
```
