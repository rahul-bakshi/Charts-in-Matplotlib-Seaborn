```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import warnings
warnings.filterwarnings('ignore')
```

```python
pd.set_option('display.max_rows', None)
df = pd.read_csv(r"C:\Users\R A H U L\Desktop\regression plot.csv")
df
```

# using regplot()

```python
plt.figure(figsize=(12,8))
sns.regplot(x="Budget", y="Sales",data=df, 
            scatter_kws = {'color':'black'}, #changing scatter color
            line_kws = {'color':'red'}) ##changing line color
plt.xlabel("Add expenditure($)")
plt.ylabel("Sales(units)")
plt.title("Effect of add expenditure on sales",fontsize=18,fontweight="bold")

plt.show()
```

# using lmplot()

```python
sns.lmplot(x="Budget", y="Sales",data=df, height=10, #used height here instead of figsize()
            scatter_kws = {'color':'green'}, #changing scatter color
            line_kws = {'color':'blue'}) #changing line color
plt.xlabel("Add expenditure($)")
plt.ylabel("Sales(units)")
plt.title("Effect of add expenditure on sales",fontsize=18,fontweight="bold")

plt.show()
```
